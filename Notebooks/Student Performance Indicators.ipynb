````python
def analyze_student_performance(data_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(data_file)

    # Display the first few rows of the DataFrame
    print("First Few Rows:")
    display(df.head())

    # Check for missing values in the DataFrame
    print("\nMissing Values:")
    display(df.isna().sum())

    # Check for duplicated rows in the DataFrame
    print("\nDuplicated Rows:")
    display(df.duplicated().sum())

    # Display the data types of each column in the DataFrame
    print("\nData Types:")
    display(df.dtypes)

    # Display the number of unique values in each column of the DataFrame
    print("\nNumber of Unique Values:")
    display(df.nunique())

    # Display the number of unique values in each column individually
    print("\nUnique Values per Column:")
    for column in df.columns:
        unique_count = df[column].nunique()
        display(f"{column}: {unique_count} unique value(s)")

    # Display the column names of the DataFrame
    print("\nColumn Names:")
    display(df.columns)

    # Display the unique categorical values in 'gender'
    print("\nUnique Categorical Values in 'gender':")
    display(df['gender'].unique())

    # Display the unique categorical values in 'lunch'
    print("\nUnique Categorical Values in 'lunch':")
    display(df['lunch'].unique())

    # Display the unique categorical values in 'parental level of education'
    print("\nUnique Categorical Values in 'parental level of education':")
    display(df['parental level of education'].unique())

    # Display the unique categorical values in 'race/ethnicity'
    print("\nUnique Categorical Values in 'race/ethnicity':")
    display(df['race/ethnicity'].unique())

    # Define numerical and categorical columns
    numerical = [feature for feature in df.columns if df[feature].dtype != 'object']
    categorical = [feature for feature in df.columns if df[feature].dtype == 'object']

    # Print the numerical and categorical features
    print("\nNumerical Features:")
    display("The length of numerical features is {} and they are: {}".format(len(numerical), numerical))

    print("\nCategorical Features:")
    display("The length of categorical features is {} and they are: {}".format(len(categorical), categorical))


# Call the function with the data file path
data_file_path = '/Users/macbook/Documents/GitHub/Student-Performance-Indicator/Data/StudentsPerformance.csv'
analyze_student_performance(data_file_path)
````

````python
def analyze_numerical_categorical_features(df):
    """
    Analyzes the numerical and categorical features of a DataFrame.

    Parameters:
        df (pandas.DataFrame): The DataFrame containing the data.

    Returns:
        numerical (list): The list of numerical features.
        categorical (list): The list of categorical features.
    """
    # Define numerical and categorical columns
    numerical = [feature for feature in df.columns if df[feature].dtype != 'object']
    categorical = [feature for feature in df.columns if df[feature].dtype == 'object']

    # Print the numerical and categorical features
    print("The length of numerical features is {} and they are: {}".format(len(numerical), numerical))
    print("The length of categorical features is {} and they are: {}".format(len(categorical), categorical))

    return numerical, categorical


#Usage:
data_file_path = '/Users/macbook/Documents/GitHub/Student-Performance-Indicator/Data/StudentsPerformance.csv'

df = pd.read_csv(data_file_path)

numerical_features, categorical_features = analyze_numerical_categorical_features(df)
````

````python
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df['average'] = df['total score']/3
df.head()

def count_students_reading_score_greater_than(df, threshold):
    """
    Counts the number of students who performed well in the reading score.

    Args:
        df (pandas.DataFrame): The DataFrame containing the student data.
        threshold (int): The threshold for considering a reading score as 'performed well'.

    Returns:
        int: The count of students who performed well in the reading score.
    """
    # Select rows where the reading score is greater than or equal to the threshold
    filtered_df = df[df['reading score'] >= threshold]
    # Count the number of values in the 'average' column that meet the condition
    count = filtered_df['average'].count()
    return count


def count_students_math_score_greater_than(df, threshold):
    """
    Counts the number of students who performed well in the math score.

    Args:
        df (pandas.DataFrame): The DataFrame containing the student data.
        threshold (int): The threshold for considering a math score as 'performed well'.

    Returns:
        int: The count of students who performed well in the math score.
    """
    # Select rows where the math score is greater than or equal to the threshold
    filtered_df = df[df['math score'] >= threshold]
    # Count the number of values in the 'average' column that meet the condition
    count = filtered_df['average'].count()
    return count


def count_students_writing_score_greater_than(df, threshold):
    """
    Counts the number of students who performed well in the writing score.

    Args:
        df (pandas.DataFrame): The DataFrame containing the student data.
        threshold (int): The threshold for considering a writing score as 'performed well'.

    Returns:
        int: The count of students who performed well in the writing score.
    """
    # Select rows where the writing score is greater than or equal to the threshold
    filtered_df = df[df['writing score'] >= threshold]
    # Count the number of values in the 'average' column that meet the condition
    count = filtered_df['average'].count()
    return count


# Calculate and display the number of students who performed well in each score category
reading_score_greater90 = count_students_reading_score_greater_than(df, 90)
math_score_greater90 = count_students_math_score_greater_than(df, 90)
writing_score_greater90 = count_students_writing_score_greater_than(df, 90)

print("Number of students who performed well in the reading score: {}".format(reading_score_greater90))
print("Number of students who performed well in the math score: {}".format(math_score_greater90))
print("Number of students who performed well in the writing score: {}".format(writing_score_greater90))
````

````python
def plot_student_performance_histograms(df):
    """
    Plots histograms of student performance based on the 'average' column, segmented by 'race/ethnicity'.

    Args:
        df (DataFrame): The input DataFrame containing student performance data.

    Returns:
        None
    """
    # Create a figure with 1 row and 3 columns, and set the size to (25, 6)
    plt.subplots(1, 3, figsize=(25, 6))

    # Subplot 1: Overall Performance
    plt.subplot(141)
    ax = sns.histplot(data=df, x='average', kde=True, hue='race/ethnicity')
    plt.title('Overall Performance')
    
    # Subplot 2: Female Performance
    plt.subplot(142)
    ax = sns.histplot(data=df[df.gender == 'female'], x='average', kde=True, hue='race/ethnicity')
    plt.title('Female Performance')

    # Subplot 3: Male Performance
    plt.subplot(143)
    ax = sns.histplot(data=df[df.gender == 'male'], x='average', kde=True, hue='race/ethnicity')
    plt.title('Male Performance')

    plt.show()
    
    
#Usage:
plot_student_performance_histograms(df)
````

````python
def plot_scores_violin(df):
    """
    Plots the violin plots for math, reading, and writing scores.

    Parameters:
        df (pandas.DataFrame): The DataFrame containing the score data.

    Returns:
        None
    """
    # Set the figure size and create subplots
    plt.figure(figsize=(18, 8))
    plt.subplot(1, 4, 1)
    plt.title('MATH SCORES')
    
    # Plot the violin plot for math scores
    sns.violinplot(y='math score', data=df, color='red', linewidth=3)
    
    plt.subplot(1, 4, 2)
    plt.title('READING SCORES')
    
    # Plot the violin plot for reading scores
    sns.violinplot(y='reading score', data=df, color='green', linewidth=3)
    
    plt.subplot(1, 4, 3)
    plt.title('WRITING SCORES')
    
    # Plot the violin plot for writing scores
    sns.violinplot(y='writing score', data=df, color='blue', linewidth=3)
    
    plt.show()


#Usage:
plot_scores_violin(df)
````

````python
def plot_gender_distribution(df):
    """
    Plots the gender distribution using a pie chart and a histogram.

    Parameters:
        df (pandas.DataFrame): The DataFrame containing the gender data.

    Returns:
        None
    """
    # Set the figure size to (8, 8)
    plt.figure(figsize=(8, 8))

    # Create subplot 1 in a 1x2 grid
    plt.subplot(1, 2, 1)

    # Plot a pie chart showing the distribution of gender using the 'gender' column from the DataFrame
    # Set the labels as 'Male' and 'Female'
    plt.pie(data=df, x=df['gender'].value_counts(), labels=['Male', 'Female'])
    plt.title('Gender Distribution')  # Add title to the subplot

    # Create subplot 2 in a 1x2 grid
    plt.subplot(1, 2, 2)

    # Plot a histogram of the gender column using seaborn's histplot function
    # Set the x-axis as the 'gender' column and use the 'hue' parameter to differentiate the bars by gender
    sns.histplot(data=df, x=df['gender'], hue='gender')
    plt.title('Gender Histogram')  # Add title to the subplot

    # Display the plot
    plt.show()


#Usage:
plot_gender_distribution(df)
````

````python
def plot_categorical_distribution(df):
    """
    Plots the categorical distribution of multiple columns in the DataFrame.

    Parameters:
        df (pandas.DataFrame): The DataFrame containing the categorical columns.

    Returns:
        None
    """
    # Create a figure with 1 row and 5 columns, and set the figure size
    fig, axes = plt.subplots(1, 5, figsize=(30, 15))

    # Plot 1: Gender
    size = df['gender'].value_counts()
    labels = ['Female', 'Male']
    colors = ['red', 'green']
    axes[0].pie(size, colors=colors, labels=labels, autopct='%.2f%%')
    axes[0].set_title('Gender', fontsize=20)  # Add title to the subplot
    axes[0].axis('off')

    # Plot 2: Race/Ethnicity
    size = df['race/ethnicity'].value_counts()
    labels = ['Group C', 'Group D', 'Group B', 'Group E', 'Group A']
    colors = ['red', 'green', 'blue', 'cyan', 'orange']
    axes[1].pie(size, colors=colors, labels=labels, autopct='%.2f%%')
    axes[1].set_title('Race/Ethnicity', fontsize=20)  # Add title to the subplot
    axes[1].axis('off')

    # Plot 3: Lunch
    size = df['lunch'].value_counts()
    labels = ['Standard', 'Free']
    colors = ['red', 'green']
    axes[2].pie(size, colors=colors, labels=labels, autopct='%.2f%%')
    axes[2].set_title('Lunch', fontsize=20)  # Add title to the subplot
    axes[2].axis('off')

    # Plot 4: Test Preparation Course
    size = df['test preparation course'].value_counts()
    labels = ['None', 'Completed']
    colors = ['red', 'green']
    axes[3].pie(size, colors=colors, labels=labels, autopct='%.2f%%')
    axes[3].set_title('Test Course', fontsize=20)  # Add title to the subplot
    axes[3].axis('off')

    # Plot 5: Parental Education
    size = df['parental level of education'].value_counts()
    labels = ['Some College', "Associate's Degree", 'High School', 'Some High School', "Bachelor's Degree", "Master's Degree"]
    colors = ['red', 'green', 'blue', 'cyan', 'orange', 'grey']
    axes[4].pie(size, colors=colors, labels=labels, autopct='%.2f%%')
    axes[4].set_title('Parental Education', fontsize=20)  # Add title to the subplot
    axes[4].axis('off')

    plt.tight_layout()
    plt.grid()

    # Display the plot
    plt.show()


# Usage:
plot_categorical_distribution(df)
````

````python
def plot_pairplot(df):
    """
    Creates a pair plot to visualize the relationship between variables in the DataFrame.

    Parameters:
        df (pandas.DataFrame): The DataFrame containing the variables.

    Returns:
        None
    """
    # Create a pair plot using seaborn's pairplot function
    # Set the 'df' DataFrame as the data source
    # Set the 'hue' parameter to 'gender' to differentiate the data points by gender
    sns.pairplot(df, hue='gender')

    plt.title('Pair Plot of Student Performance by Gender')  # Add title to the plot

    plt.show()


# Example usage:
plot_pairplot(df)
````
